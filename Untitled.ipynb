```python
dict_order = {2:2,
 1:3,
 11:1,
 13:3,
  7:3,
  8:1,
  12:1,
   14:2,
   4:1, 
  10:1,
   6:1,
   3:3,
   15:2,
    9:2,
    5:1}

dict_category = {
    2:'shoulders',
    1:'arms',
    11:'legs',
    13:'arms',
    7: 'legs',
    8: 'legs',
    12: 'back',
    14: 'abs',
    4: 'chest',
    10: 'legs',
    15: 'legs',
    6: 'abs',
    3: 'legs',
    9: 'shoulders',
    5: 'arms'
}


dict_name_category = {
    'shoulders':13,
    'legs':9,
    'chest':11,
    'back':12,
    'arms':8,
    'abs':10
    
}

def get_muscle_importance(id_):
    
    return dict_order.get(id_,3)

def get_muscle_category(id_):
    
    return dict_category.get(id_,'all')

def get_muscle_category_id(category):
    
    return dict_name_category.get(category,'all')
```

```python
df_muscle['importance'] = df_muscle['id'].apply(get_muscle_importance) 
df_muscle['category'] = df_muscle['id'].apply(get_muscle_category)
df_muscle['category_id'] = df_muscle['category'].apply(get_muscle_category_id)
```
